```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json

def blkom_stream_ext(link):
  url = link
  lst = []
  res = []
  ll = []

  page = requests.get(url)
  soup = BeautifulSoup(page.content,'html.parser')

  for i in soup.find('div',{'class': 'servers col-xs-12 col-md-9 pull-left'}):
    if i != ' ':
      #lst.append(str(i).rstrip())
      soup2 = BeautifulSoup(str(i).rstrip(),'html.parser')
      lst.append(soup2.find_all('a'))
      res = [ele for ele in lst if ele != []]

  lst = res[0]

  for i in lst:
    string = str(i)
    link = string[ string.find('"')+1 : string.rfind('"') ]
    server = string[string.find('>') + 1 : string.rfind('<')]
    data = {server:link}
    ll.append(data)

  return ll


def blkom_eps(url):
  link = url
  page = requests.get(url)
  soup = BeautifulSoup(page.content,'html.parser')
  lst = []
  eps = {}

  for i in soup.find_all('li' , {'class' : 'episode-link'}):
    link = ""+i.a.get('href')
    link = link.rstrip()
    lst.append(link)
  index = 1
  for i in range((len(lst))):
    eps[index] = lst[i]
    index = index + 1
  return eps



def blkom_leak_anime(url):
  #for ex: https://www.egyanime.com/anime?do=1243
  x = blkom_eps(url)
  lst = []

  if x == {}:
    stream = blkom_stream_ext(url)
    data = {
        1 : stream
    }
    lst.append(data)

  for key, value in x.items() :
    stream_ext = blkom_stream_ext(value)
    if key is not None and stream_ext is not None:
      data = {
          key: stream_ext
      }
      lst.append(data)

  return lst
```

```python
def leak_anime_list(target):
  url = target
  page = requests.get(url)
  soup = BeautifulSoup(page.content,'html.parser')
  lst = []
  rate_lst = []
  index_rate = 0
  for x in soup.find_all('div',{'class':'badge dark-red'}):
    rate_lst.append(x.get_text())

  for i in soup.find_all('div',{'class':'content-inner'}):
    link = ""+i.a.get('href')
    link = link.rstrip()

    img = "https://blkom.com"+i.img.get("data-original")
    img = img.rstrip()

    des = i.p.text
    des = des.rstrip()

    name = str(i.a.get('href'))
    index = name.rfind("/")
    name = name[index+1:]

    eps = blkom_leak_anime(link)

    data = {
        'name':name,
        'rate':rate_lst[index_rate],
        'src-img': img,
        'description': des,
        'episodes': eps
    }
    lst.append(data)
    index_rate = index_rate + 1
  return lst

```

```python
for i in range(2):
    link = leak_anime_list(f"https://animeblkom.net/series-list?states=finished&year=2007-2021&rate=9-10&age=G-R&page={i}")
    with open(f"data-series{i}.json", 'w') as output:
        json.dump(link, output)
```

```python
def extract_streams_from_blkom(link):

    """
    Extracts streaming links for a given URL.

    Parameters:
    url (str): The URL to extract streaming links from.

    Returns:
    list: A list of dictionaries containing the name of each server and its corresponding link.
    """

    url = link
    ll = []

    page = requests.get(url)
    soup = BeautifulSoup(page.content,'html.parser')

    for i in soup.find('div',{'class': 'servers col-xs-12 col-md-9 pull-left'}):
        if i != ' ':
            soup2 = BeautifulSoup(str(i).rstrip(),'html.parser')
            lst = soup2.find_all('a')
            for link_tag in lst:
                link = link_tag.get('data-src')
                server = link_tag.string
                data = {server: link}
                ll.append(data)

    return ll

def extract_episodes_from_blkom(url):
    """
    Extracts the episode links for a given URL.

    Parameters:
    url (str): The URL to extract episode links from.

    Returns:
    dict: A dictionary containing the index number of each episode and its corresponding link.
    """
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    episode_links = {}

    for index, episode in enumerate(soup.find_all('li', {'class': 'episode-link'}), start=1):
        link = 'https://blkom.com' + episode.a['href']
        episode_links[index] = link

    return episode_links


def extract_streams_episodes(url):
    """
    Extracts the streaming links and episode links for a given URL from blkom.com.

    Parameters:
        url (str): The URL to extract links from.

    Returns:
        list: A list of dictionaries containing the episode index and its corresponding streaming links.
    """

    episode_links = extract_episodes_from_blkom(url)
    lst = []

    if episode_links == {}:
        stream = extract_streams_from_blkom(url)
        data = {1: stream}
        lst.append(data)
    else:
        for index, episode_url in episode_links.items():
            stream_ext = extract_streams_from_blkom(episode_url)
            if index is not None and stream_ext is not None:
                data = {index: stream_ext}
                lst.append(data)

    return lst


def leak_anime_list(target):
    """
    Scrapes anime information from the given target URL.

    Parameters:
    target (str): The URL of the webpage containing anime information.

    Returns:
    list: A list of dictionaries, where each dictionary contains the name, rating, image source,
          description, and episodes of an anime.
    """
    url = target
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    anime_list = []
    rating_list = []

    for rating_elem in soup.find_all('div', {'class': 'badge dark-red'}):
        rating_list.append(rating_elem.get_text())

    for anime_elem in soup.find_all('div', {'class': 'content-inner'}):
        anime_link = "https://blkom.com" + anime_elem.a.get('href').rstrip()
        anime_image = "https://blkom.com" + anime_elem.img.get("data-original").rstrip()
        anime_description = anime_elem.p.text.rstrip()
        anime_name = anime_elem.a.get('href').split('/')[-1]
        anime_episodes = extract_streams_episodes(anime_link)
        anime_data = {
            'name': anime_name,
            'rate': rating_list.pop(0),
            'src-img': anime_image,
            'description': anime_description,
            'episodes': anime_episodes
        }
        anime_list.append(anime_data)

    return anime_list


for i in range(1):
    link = leak_anime_list(f"https://animeblkom.net/series-list?states=finished&year=2007-2021&rate=9-10&age=G-R&page={i}")
    with open(f"data-series{i}.json", 'w') as output:
        output.write(json.dumps(link))
```

```python
import concurrent.futures
import requests
from bs4 import BeautifulSoup

def extract_streams_from_blkom(link, session):
    """
    Extracts streaming links for a given URL.

    Parameters:
        url (str): The URL to extract streaming links from.
        session (requests.Session): The session object to use for the request.

    Returns:
        list: A list of dictionaries containing the name of each server and its corresponding link.
    """

    page = session.get(link)
    soup = BeautifulSoup(page.content, 'lxml')

    servers = soup.find('div', {'class': 'servers col-xs-12 col-md-9 pull-left'})
    if not servers:
        return []

    stream_list = []
    for link_tag in servers.find_all('a'):
        server = link_tag.string
        link = link_tag.get('data-src')
        data = {server: link}
        stream_list.append(data)

    return stream_list


def extract_episodes_from_blkom(url, session):
    """
    Extracts the episode links for a given URL.

    Parameters:
        url (str): The URL to extract episode links from.
        session (requests.Session): The session object to use for the request.

    Returns:
        dict: A dictionary containing the index number of each episode and its corresponding link.
    """
    page = session.get(url)
    soup = BeautifulSoup(page.content, 'lxml')

    episode_links = {}
    for index, episode in enumerate(soup.find_all('li', {'class': 'episode-link'}), start=1):
        link = 'https://blkom.com' + episode.a['href']
        episode_links[index] = link

    return episode_links


def extract_streams_episodes(url, session):
    """
    Extracts the streaming links and episode links for a given URL from blkom.com.

    Parameters:
        url (str): The URL to extract links from.
        session (requests.Session): The session object to use for the request.

    Returns:
        list: A list of dictionaries containing the episode index and its corresponding streaming links.
    """

    episode_links = extract_episodes_from_blkom(url, session=session)
    episode_list = []

    if not episode_links:
        stream_list = extract_streams_from_blkom(url, session=session)
        episode_data = {1: stream_list}
        episode_list.append(episode_data)
    else:
        with concurrent.futures.ThreadPoolExecutor() as executor:
            future_to_episode = {executor.submit(extract_streams_from_blkom, episode_url, session=session): index for index, episode_url in episode_links.items()}
            for future in concurrent.futures.as_completed(future_to_episode):
                index = future_to_episode[future]
                stream_list = future.result()
                if index is not None and stream_list:
                    episode_data = {index: stream_list}
                    episode_list.append(episode_data)

    return episode_list


def leak_anime_list(target, session):
    """
    Scrapes anime information from the given target URL.

    Parameters:
    target (str): The URL of the webpage containing anime information.
    session (requests.Session): The session object to use for the request.

    Returns:
    list: A list of dictionaries, where each dictionary contains the name, rating, image source,
          description, and episodes of an anime.
    """
    page = session.get(target)
    soup = BeautifulSoup(page.content, 'lxml')

    anime_list = []
    rating_list = [rating_elem.get_text() for rating_elem in soup.find_all('div', {'class': 'badge dark-red'})]

    for anime_elem in soup.find_all('div', {'class': 'content-inner'}):
        anime_link = "https://blkom.com" + anime_elem.a['href'].rstrip()
        anime_image = "https://blkom.com" + anime_elem.img['data-original'].rstrip()
        anime_description = anime_elem.p.text.rstrip()
        anime_name = anime_elem.a['href'].split('/')[-1]
        anime_episodes = extract_streams_episodes(anime_link, session)
        anime_data = {
            'name': anime_name,
            'rate': rating_list.pop(0),
            'src-img': anime_image,
            'description': anime_description,
            'episodes': anime_episodes
        }
        anime_list.append(anime_data)

    return anime_list



```
